**Importing Libraries**

```python
import numpy as np
import pandas as pd
```

**Merging the two given datasets**

```python
df = pd.read_csv('AWCustomers.csv')
extra = pd.read_csv('AWSales.csv')
df = pd.merge(df,extra)
df.head()
```

**Dropping duplicate ids present if any**

```python
df = df.drop_duplicates(subset=['CustomerID'])
```

**Removing the extra columns**

```python
df = df.drop(['CustomerID','Title','FirstName','MiddleName','LastName','Suffix','AddressLine1','AddressLine2','PostalCode','PhoneNumber','LastUpdated','City','StateProvinceName'],axis = 1)
df.info()
```

```python
df.head()
```

**Calculating Age from DOB**

```python
from datetime import datetime
from datetime import date
def calculate_age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['age'] = df['BirthDate'].apply(calculate_age)
df = df.drop('BirthDate',axis = 1)
df.head()
```

**Checking for NULL values**

```python
df.isnull().sum()
```

**Converting into numerical form for further processing**

*   **One Hot Encoding**




```python
Genderr = pd.get_dummies(df.Gender)
Genderr = Genderr.drop('F',axis = 1)
df = pd.concat([df,Genderr],axis = 'columns')
df = df.drop('Gender',axis = 'columns')
df.rename(columns = {'M':'Male'},inplace = True)
df.head()
```

```python
Status = pd.get_dummies(df.MaritalStatus,drop_first=True)
df = pd.concat([df,Status],axis = 'columns')
df = df.drop('MaritalStatus',axis = 1)
df.head()
```

**Mapping the Education with labels More the education More is the weightage Similar with Occupation**

```python
df['Occupation']=df['Occupation'].map({'Manual':1,'Skilled Manual':2,'Clerical':3,'Management':4,'Professional':5})
df['Education']=df['Education'].map({'Partial High School':1,'High School':2,'Partial College':3,'Bachelors':4,'Graduate Degree':5})
df.head()
```

**PERFORMING ONE HOT ENCODING**

```python
CountryRegionName = pd.get_dummies(df.CountryRegionName,drop_first=True)
df = pd.concat([df,CountryRegionName],axis = 'columns')
df = df.drop('CountryRegionName',axis = 1)
df.head()
```

```python
df['4 bins'] = pd.qcut(df['age'], q=4)
df.head()
```

**SCALING AND NORMALIZING THE DATA**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled=scaler.fit_transform(df[['YearlyIncome','age','AvgMonthSpend']])
df['YearlyIncome_scaled']=scaled[:,0]
df['Age_scaled']=scaled[:,1]
df['AvgMonthSpend_Scaled']=scaled[:,2]
df.drop(['YearlyIncome','age'],axis=1,inplace=True)
df = df.drop('AvgMonthSpend',axis=1)
df.head()
```

**COSINE**

```python
from numpy.linalg import norm
A = df['YearlyIncome_scaled']
B = df['Education']
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("Cosine Similarity between YearlyIncome_Scaled and Education", cosine)
```

```python
from scipy.spatial import distance
distance.cosine(df['Education'].values,df['YearlyIncome_scaled'].values)
distance.cosine(df['Education'].values,df['AvgMonthSpend_Scaled'].values)
distance.cosine(df['TotalChildren'].values,df['BikeBuyer'].values)
```

**JACCARD SCORE**

```python
from sklearn.metrics import jaccard_score
print(jaccard_score(df['Canada'], df['Male']))
print(jaccard_score(df['Germany'], df['Male']))
```

**PEARSON CORRELATION**

```python
print(np.corrcoef(df['Education'], df['YearlyIncome_scaled']))
print(np.corrcoef(df['BikeBuyer'], df['TotalChildren']))
print(np.corrcoef(df['BikeBuyer'], df['Age_scaled']))
print(np.corrcoef(df['Age_scaled'], df['TotalChildren']))
```
