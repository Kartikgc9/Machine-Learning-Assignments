```python
from pandas import set_option
set_option('display.width',100)
set_option('precision',3)
```

```python
correlation= iris_data.drop('Id',axis=1).corr(method="pearson")
correlation  
```

```python
fig, ax = plt.subplots(figsize =(5, 4))
sns.heatmap(correlation, ax = ax)
```

```python
class_counts = iris_data.groupby('Species').size()
print(class_counts)
```

```python
skew = iris_data.skew()
print(skew)
```

**Different types of plots**

```python
from matplotlib import pyplot
iris_data.drop('Id',axis=1).hist(figsize = (8,8))
pyplot.show()
```

```python
iris_data.drop('Id',axis=1).plot(kind = 'density',subplots = True, layout = (3,3), sharex = False,figsize = (10,10))
pyplot.show()
```

```python
iris_data.drop('Id',axis=1).plot(kind = 'box',subplots = True, layout = (3,3),sharex = False, sharey = False,figsize = (10,10))
```

```python
iris_data.drop('Id',axis=1).boxplot(by = "Species", figsize=(8,8))
plt.show()
```

```python
sns.pairplot(iris_data.drop('Id',axis=1), hue="Species")
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris_data.drop('Id',axis=1), figsize = (8,8))
```
